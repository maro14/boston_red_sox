```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from pandas import DataFrame
from collections import Counter
```

```python
url = 'https://www.baseball-reference.com/players/o/ortizda01.shtml#batting_standard::none' # API from website
ortiz = pd.read_html(url, header=0,)[0]
ortiz
```

```python
ortiz.tail()
```

```python
x = ortiz.Tm

play_counts = Counter(x)
df = pd.DataFrame.from_dict(play_counts, orient='index')
df =df[0].sort_values(ascending = True)[5:20]

df.plot(kind='bar')
plt.xlabel('Team')
plt.ylabel('Play for Team')
plt.title('David Ortiz play for')


plt.show()
```

```python
raw_data = {'Name': ['David Ortiz', 'David Ortiz', 'David Ortiz', 'David Ortiz', 'David Ortiz', 'David Ortiz'],
           'Games': [11, 67, 89, 122, 200, 165],
           'Homerun': [2, 12, 18, 21, 43, 19]}
data1 = pd.DataFrame(raw_data)
data1
```

```python
data2 = pd.read_csv('david.csv') # API from data
data2
```

```python
sns.set(style='darkgrid')
baseball = data2.groupby('Tm').sum()
baseball = baseball.sort_values(by = 'HR', ascending = False)[0:100]

baseball['HR'].plot(kind='bar')
plt.xlabel('Team')
plt.ylabel('Homerun')
plt.title('David Ortiz Homerun')

plt.show()
```

```python
# Set style of scatterplot
sns.set_context("notebook", font_scale=1.1)
sns.set_style("ticks")

# Create scatterplot of dataframe
sns.lmplot('G', # Horizontal axis
           'HR', # Vertical axis
           data=data2, # Data source
           fit_reg=False, # Don't fix a regression line
           hue="Tm", # Set color
           scatter_kws={"marker": "D", # Set marker style
                        "s": 100}) # S marker size

# Set title
plt.title('Histogram of David Ortiz Homerun per Games')

# Set x-axis label
plt.xlabel('Games')

# Set y-axis label
plt.ylabel('Homerun')
```

```python
# Set style of scatterplot
sns.set_context("notebook", font_scale=1.1)
sns.set_style("ticks")

# Create scatterplot of dataframe
sns.lmplot('Year', # Horizontal axis
           'HR', # Vertical axis
           data=data2, # Data source
           fit_reg=False, # Don't fix a regression line
           hue="Tm", # Set color
           scatter_kws={"marker": "D", # Set marker style
                        "s": 100}) # S marker size

# Set title
plt.title('Histogram of David Ortiz Homerun per Years')

# Set x-axis label
plt.xlabel('Years')

# Set y-axis label
plt.ylabel('Homerun')
```

```python
sns.set_context('notebook', font_scale=1.1)
sns.set_style('ticks')

sns.lmplot('Year', 
           'SLG', 
           data=data2,
           hue="Tm", 
           scatter_kws={"marker": "D", "s": 120})

plt.title("Ortiz Slugging")
plt.xlabel('Years')
plt.ylabel('Slugging')
```
