# Boston Red Sox picthers

```python
red = pd.read_csv('red_sox_pitcher.csv') # API from Local data
red
```

### How much red sox pitcher got saves game for the team

```python
red[['Name', 'SV']]
```

```python
red.groupby('Name').sum().plot(kind='bar', y='SV',color="g" ) # Pandas output
```

```python
closer = red.groupby('SV')
closer = closer.sum()
closer = closer.sort_values(['SV'], ascending=False).head()# Top closer Pitcher with higher save

closer
```
